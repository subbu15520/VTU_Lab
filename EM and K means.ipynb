```python
import matplotlib.pyplot as plot
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.mixture import GaussianMixture
import numpy as np
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
data = pd.DataFrame(iris.data)
data.columns = ['Sepal_Lenght','Sepal_Width','Petal_Lenght','Petal_Width']
print(data)
```

```python
targets = pd.DataFrame(iris.target)
targets.columns = ['targets']
print(targets)
```

```python
colorMap = np.array(['red','green','blue','yellow'])
```

```python
#real clusters
plot.scatter(data.Petal_Lenght, data.Petal_Width, c = colorMap[targets.targets], s = 5)
plot.title('real clusters')
plot.xlabel('Petal lenght')
plot.ylabel('Petal Width')
```

```python
#K means clustering
model = KMeans(n_clusters = 4)
model.fit(data)

plot.scatter(data.Petal_Lenght, data.Petal_Width, c = colorMap[model.labels_], s = 5)
plot.title('K Means clustering')
plot.xlabel('Petal lenght')
plot.ylabel('Petal Width')
```

```python
#Gaussian mixture odel clustering
scaler = preprocessing.StandardScaler()
scaledData = pd.DataFrame(scaler.fit(data).transform(data), columns = data.columns)

gmm = GaussianMixture(n_components = 4)
gmm.fit(scaledData)
gmm_targets = gmm.predict(scaledData) 

plot.scatter(data.Petal_Lenght, data.Petal_Width, c = colorMap[gmm_targets], s = 5)
plot.title('GMM clustering')
plot.xlabel('Petal lenght')
plot.ylabel('Petal Width')
```

```python
print("Observation:The GMM Clustering using EM Algorithm matched the True Labels more closely than the K Means Clustering")
```
